```python
def get_embedding(text: str):
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text,
    )

    embedding = response.data[0].embedding
    return embedding
```

```python
for fact in facts:
    fact_id = fact["id"]
    text = fact["content"]
    print(f"{fact_id}: {text}")

    embedding = get_embedding(text)
    
    cur = conn.cursor()

    cur.execute("""
    INSERT INTO document_chunks
    (document_id, chunk_text, embedding)
    VALUES (%s, %s,%s)
    """,
    (fact_id, text, embedding))

    conn.commit()
```

```python
question = "Are you from mars?"
question_embedding = Vector(get_embedding(question))

cur.execute("""
SELECT
    chunk_text,
    embedding <=> %s::vector AS distance
FROM document_chunks
ORDER BY distance;
WHERE distance 
""", (question_embedding,))

rows = cur.fetchall()

context = "\n\n".join(
    f"{content} score: {score}"
    for content, score in rows
)

print(context)
```

```python
prompt = f"""
You are a helpful customer support assistant.

Answer ONLY using the information below.

If the answer is not present, say you don't know.

Knowledge Base

{context}

Question:
{question}
"""

response = client.responses.create(
    model="gpt-4.1",
    input=prompt,
)

print("Answer:\n")
print(response.output_text)
```
